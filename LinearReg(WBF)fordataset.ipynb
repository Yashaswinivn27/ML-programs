```python
import pandas as pd
import numpy as np
import math
import operator
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('headbrain.csv')
print(data.head())
```

```python
X=data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values
```

```python
plt.scatter(X,Y)
```

```python
meanx=np.mean(X)
meany=np.mean(Y)
print(meanx,meany)
```

```python
# now let us find the slope and intercept
m=len(X)
numerator=0
denominator=0
for i in range(m):
    numerator+=(X[i]-meanx)*(Y[i]-meany)
    denominator+=(X[i]-meanx)**2
m=numerator/denominator
c=meany-(m*meanx)
print("The slope is --->",m)
print("The intercept is--->",c)
```

```python
# Plotting the scatter plot of data points
plt.scatter(X, Y, color='blue', label='Data Points')

# Plotting the regression line
regression_line = m*X + c
plt.plot(X, regression_line, color='red', label='Regression Line')

# Adding labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()

# Displaying the plot
plt.show()

```

```python
# now let us calculate goodness of fit of our model
d=0
n=0
for i in range(5):
    ypred=c+m*X[i]
    n+=(ypred-Y[i])
    d+=(meany-Y[i])
result=1-(n/d)
print(result)
    
```
